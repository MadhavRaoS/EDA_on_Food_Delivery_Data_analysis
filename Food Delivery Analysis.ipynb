# Data Cleaning


## Lets Drop Duplicates

```python
ds.drop_duplicates(inplace = True)
ds.shape
```

```python
ds.head()
```

##  Lets Clean rate column

```python
ds['rate'].unique()
```


## Lets remove Strings and "/5" in rate column

```python
def apple(a):
    if(a=='NEW' or a=='-'):
        return np.nan
    else:
        a=str(a).split('/')
        a=a[0]
        return float(a)   
ds['rate'] = ds['rate'].apply(apple)
ds['rate'].head()
```

```python
ds['rate'].fillna(ds['rate'].mean(),inplace=True)
```

```python
ds['rate'].isnull().sum()
```

```python
ds.info()
```

```python
ds.head()
```

```python
ds.dropna(inplace=True)
ds.head()
```

```python
ds['location'].unique()
```

```python
ds['listed_in(city)'].unique()
```


## Lets keep only one in location or listed_in(city)

### Removing listed_in(city) column

```python
ds=ds.drop(['listed_in(city)'],axis=1)
ds.head()
```

```python
ds['location'].unique()
```

```python
location = ds['location'].value_counts(ascending=False)
location
```

```python
location_lessthan200=location[location<200]
location_lessthan200
```

## Cleaning location column

```python
def loc(b):
    if(b in location_lessthan200):
        return 'others'
    else:
        return b
        
ds['location']=ds['location'].apply(loc)
ds['location'].value_counts()
```

```python
ds.head()
```

## Lets Clean rest_type

```python
rest_types = ds['rest_type'].value_counts(ascending  = True)
rest_types_lessthan900 = rest_types[rest_types<900]
def rest(a):
    if(a in rest_types_lessthan900):
        return 'others'
    else:
        return a
        
ds['rest_type']=ds['rest_type'].apply(rest)
ds['rest_type'].value_counts()
```

```python
ds.head()
```

## Lets clean cuisines 

```python
cuisines=ds['cuisines'].value_counts(ascending = True)
cuisines_lessthan150=cuisines[cuisines<150]
def rolex(a):
    if(a in cuisines_lessthan150):
        return 'others'
    else:
        return a
ds['cuisines']=ds['cuisines'].apply(rolex)
ds['cuisines'].value_counts()
```

```python
ds.head()
```

### Q.Which locations has highest Votes?

### Q. What are the top cuisines ?
